Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.


```python
car.isnull().sum()
```

```python
car.select_dtypes(include=['number'])
```

```python
car.select_dtypes(include=['number']).columns
```

```python
numeric_cols = car.select_dtypes(include=['number']).columns
car[numeric_cols] = car[numeric_cols].fillna(car[numeric_cols].mean())
```

```python
car.isnull().sum()
```

```python
str_cols = car.select_dtypes(exclude = ['number']).columns
car[str_cols] = car[str_cols].fillna("Unknown")
car.isnull().sum()
```
